```python
from bs4 import BeautifulSoup
import requests
from time import sleep
import pandas as pd
```

```python
html = requests.get("https://wordpress.org/support/plugin/litespeed-cache/reviews/").text
soup = BeautifulSoup(html, 'html5lib')
page_numbers = soup('a', 'page-numbers')
NUM_PAGES = max([int(a.text) for a in page_numbers if a.text.isnumeric()])
```

```python
NUM_PAGES
```

```python
%%time
print(f'Total Page Number: {NUM_PAGES}')
      
base_url = 'https://wordpress.org/support/plugin/litespeed-cache/reviews/page/'

reviews = pd.DataFrame(columns=['topic', 'hyperlink', 'author', 'rating'])

for page_num in range(1, NUM_PAGES+1):
    print(f'souping page {page_num}, {reviews.shape[0]} reviews found so far')
    url = base_url + str(page_num)
    
    soup = BeautifulSoup(requests.get(url).text, 'html5lib')
    
    for li in soup('li', 'bbp-topic-title'):
        try:
            topic = li.find('a', 'bbp-topic-permalink').text
            hyperlink = li.find('a', 'bbp-topic-permalink').get('href')
            author = li.find('span', 'bbp-author-name').text
            rating = li.find('div', 'wporg-ratings').get('title')

            reviews = reviews.append({'topic': topic, 'hyperlink': hyperlink, 'author': author, 'rating': rating}, ignore_index=True)
        except:
            continue

print(f'{reviews.shape[0]} reviews\' info scraped except for content')
```

```python
reviews
```

```python
%%time
for link in reviews['hyperlink']:
    soup = BeautifulSoup(requests.get(link).text, 'html5lib')
    
    content = ""
    for div in soup('div', 'bbp-topic-content'):
        try:
            content += div.p.text
        except:
            continue
    reviews.loc[reviews['hyperlink']==link, 'content'] = content
```

```python
from datetime import date
reviews['rating'] = reviews['rating'].map(lambda x: int(x[0]))
reviews['last update'] = date.today().strftime('%Y%m%d')
reviews.drop('hyperlink', axis = 1, inplace=True)
reviews
```

## Connect to Postgre DB on aws

```python
connect_info = {}
with open('db_connection_info.txt', 'r') as file:
    for line in file:
        info = line.split()
        if len(info)==2:
            connect_info[info[0]] = info[1]

username = connect_info['username']
password = connect_info['password']
endpoint = connect_info['endpoint']
port = connect_info['port']
```

```python
%%time
import sqlalchemy as sq


engine = sq.create_engine('postgresql://'+username+':'+password+'@'+endpoint+':'+port+'/postgres')
connection = engine.connect()
# class Database():
#     # replace the user, password, hostname and database according to your configuration according to your information
#     engine = db.create_engine('postgresql://user:password@hostname/database_name')
#     def __init__(self):
#         self.connection = self.engine.connect()
#         print("DB Instance created")

reviews.to_sql('reviews', engine, if_exists='replace') #append
```
